```python
import pandas as pd
df1 = pd.read_json('../_zap/dados/zap.json')
df2 = pd.read_json('../_quinto/dados/quinto.json')
df3 = pd.read_json('../_viva/dados/viva.json')
df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df
```

```python
#Removendo os valores que tem TODOS os dados iguais 
df = df.drop_duplicates(subset=['price' ,'address','rooms','parking','toilets','size', 'penthouse'])
df
```

```python
#Verificando se tem dados duplicados
df.duplicated(subset=['price' ,'address','rooms','parking','toilets','size', 'penthouse']).value_counts()
```

```python
df.to_json('all_aps.json', orient='records', force_ascii=False, indent=4)
```
